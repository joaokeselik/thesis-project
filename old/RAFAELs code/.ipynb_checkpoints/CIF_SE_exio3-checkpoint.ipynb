```python
exio3.calc_all()
```

```python
cf = pd.read_csv('/Users/rafa/Box/Python/Cha_Fac_footprints_RL.csv',';', header=[0], index_col = [0])
cf
```

```python
# Consumption matrix
#M.EUR
Y = exio3.Y.SE.groupby(level=1, sort=False).sum()[['Final consumption expenditure by households']]
Y
```

```python
# Multiply the matrix comsumption-based accounting by the cf (I'm doing it only for CO2eq.)
# You can use Richad Wood's file

# Impact of total comsumption per category
# Impact per M.EUR (total consumption)

cba_SE = exio3.satellite.D_cba.SE.T.dot(cf.T)[['Carbon Footprint [kg CO2-eq.]']]
cba_SE
```

```python
# Divide the matrix total consumption (Y) by the matrix impact of total consumption (cba_SE) to find the consumption intensity factors

# CO2 per M.EUR

co2_per_ME = cba_SE[['Carbon Footprint [kg CO2-eq.]']].div(Y['Final consumption expenditure by households'], axis=0)
co2_per_ME
```

```python
# Merging the two matrices
multi = Y.merge(co2_per_ME, on='sector')
multi
```

```python
#import numpy as np
#np.average(multi["Carbon Footprint [kg CO2-eq.]"], weights=multi["Final consumption expenditure by households"])
```

```python
# canculating the weighted average
#(df["quantity"] * df["weight"]).sum() / df["weight"].sum()
# kg per euros
wavg_kg_EUR = (multi["Carbon Footprint [kg CO2-eq.]"]*multi["Final consumption expenditure by households"]).sum()/multi["Final consumption expenditure by households"].sum()/1000000
wavg_kg_EUR
```

```python
#alternatevely
(co2_per_ME["Carbon Footprint [kg CO2-eq.]"]*Y["Final consumption expenditure by households"]).sum()/Y["Final consumption expenditure by households"].sum()/1000000
```

```python
# g per SEK
#still need to ajust for inflation
wavg_g_SEK = wavg_kg_EUR*100
wavg_g_SEK
```
