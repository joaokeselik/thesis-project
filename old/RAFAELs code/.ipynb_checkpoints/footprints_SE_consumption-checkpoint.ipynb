# Footprints per € spent in each consumption category for Sweden

```python
import pymrio
import pandas as pd
import numpy as np
```

```python
#download the EXIOBASE 3.4 - IOT - 2011 - pxp
#from on: https://exiobase.eu/index.php/data-download/exiobase3mon
#import database
exio3 = pymrio.parse_exiobase3('C:\\Users\\keseljoa\\Desktop\\IOT\\IOT_2011_pxp')

#see https://pymrio.readthedocs.io/en/latest/index.html
```

```python
#calculate missing dataframes
exio3.calc_all()
```

```python
#chatacterisation factors to convert envrionemtal requirements into footprints
characterisation = pd.read_csv('Cha_Fac_footprints_RL.csv',';', header=[0], index_col = [0])
characterisation
```

```python
#total environmental or social requirements per consumptin category for Sweden
#unit: requirement per M.EUR
## Ex.:
## emissions e.g. CO2 - combustion - air: kg per M.EUR
## land use: km2 per M.EUR
## water use: Mm3 per M.EUR
## energy use: TJ per M.EUR

D_cba_SE = exio3.satellite.D_cba.SE.T
D_cba_SE

#O objeto D_cba contain consumption-based accounts
```

```python
#unit: footprints per M.EUR
D_cba_SE_footprints = D_cba_SE.dot(characterisation.T)
D_cba_SE_footprints
```

```python
D_cba_SE_footprints.to_csv('footprints_SE.csv')
D_cba_SE_footprints.to_excel(r'footprints_SE_excel.xlsx')
```
