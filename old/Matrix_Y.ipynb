```python
#total environmental or social requirements per consumption category for Sweden
D_cba_SE = exio3.satellite.D_cba.SE.T
```

```python
#chatacterisation factors to convert environmemtal requirements into footprints
characterisation  = pd.read_csv('EXIOBASE341f_CC500f.csv',',', header=[0], index_col = [0])
```

```python
#keeping only stressors in commmon
df1 = characterisation.copy()
df2 = D_cba_SE.copy()

characterisation = df1.drop([idx for idx in df1.index if idx in df1.index and idx not in df2.columns], axis=0)
D_cba_SE = df2.drop([col for col in df2.columns if col in df2.columns and col not in df1.index], axis=1)
```

```python
D_cba_SE_footprints = D_cba_SE.dot(characterisation)
```

```python
import functools
def add(x,y): return x+y

emissions = D_cba_SE_footprints['GHG emissions (GWP100)|kg CO2 eq.|Problem oriented approach: baseline (CML, 2001)|GWP100 (IPCC, 2007)'].tolist()
emissions_total = functools.reduce(add, emissions)
emissions_total
```

```python
Y_SE = exio3.Y.SE.groupby(level=1, sort=False).sum()
```

```python
consumption_households_list = Y_SE['Final consumption expenditure by households'].tolist()
consumption_households = functools.reduce(add, consumption_households_list)

consumption_ngos_list = Y_SE['Final consumption expenditure by non-profit organisations serving households (NPISH)'].tolist()
consumption_ngos = functools.reduce(add, consumption_ngos_list)

consumption_government_list = Y_SE['Final consumption expenditure by government'].tolist()
consumption_government = functools.reduce(add, consumption_government_list)

consumption_total = consumption_households + consumption_ngos + consumption_government
consumption_total_SEK = consumption_total*INFLATION_EUR_2011_2020*EUR_TO_SEK*M_EUR
consumption_total_SEK
```

```python
emissions_per_SEK = emissions_total/consumption_total_SEK
print(str(emissions_per_SEK) + ' [Kg CO\u2082eq]/[SEK]')
```
